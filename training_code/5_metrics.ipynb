# Experimental Evaluation and Performance Assessment

## Dependency Installation and Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    precision_score, 
    recall_score, 
    f1_score, 
    fbeta_score, 
    roc_auc_score, 
    precision_recall_curve, 
    auc, 
    confusion_matrix
)
```

## Classification Quality Evaluation

```python
def calculate_metrics_from_counts(tp, fp, fn, tn):
    # Total count
    total = tp + fp + fn + tn
    # Accuracy (share of correct predictions)
    accuracy = (tp + tn) / (tp + fp + fn + tn)
    
    # Precision (Precision)
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    
    # Recall (Recall)
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    
    # F1-Score
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    # F2-Score
    beta = 2
    f2 = (1 + beta**2) * (precision * recall) / ((beta**2 * precision) + recall) if ((beta**2 * precision) + recall) > 0 else 0
    
    # Specificity (True Negative Rate)
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
    
    # ROC-AUC (Simplified one-point calculation)
    roc_auc = (recall + specificity) / 2
    
    # PR-AUC (Simplified one-point calculation)
    pr_auc = precision * recall 

    # Print results
    print(f"--- Calculation Results ---")
    print(f"Matrix: TP={tp}, FP={fp}, FN={fn}, TN={tn}")
    print(f"---------------------------")
    print(f"Accuracy:  {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1-Score:  {f1:.4f}")
    print(f"F2-Score:  {f2:.4f}")
    print(f"ROC-AUC:   {roc_auc:.4f}")
    print(f"PR-AUC:    {pr_auc:.4f}")
    
    return {
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "f2": f2,
        "roc_auc": roc_auc,
        "pr_auc": pr_auc
    }
```
